```python
import os, shutil, random

source = "../dataset/original"
train = "../dataset/train"
val = "../dataset/val"
test = "../dataset/test"

random.seed(42)

# Create class folders automatically
for cls in os.listdir(source):
    os.makedirs(os.path.join(train, cls), exist_ok=True)
    os.makedirs(os.path.join(val, cls), exist_ok=True)
    os.makedirs(os.path.join(test, cls), exist_ok=True)

# Split images
for cls in os.listdir(source):
    images = os.listdir(os.path.join(source, cls))
    random.shuffle(images)

    total = len(images)
    train_end = int(0.7 * total)
    val_end = int(0.85 * total)

    for img in images[:train_end]:
        shutil.copy2(
            os.path.join(source, cls, img),
            os.path.join(train, cls, img)
        )

    for img in images[train_end:val_end]:
        shutil.copy2(
            os.path.join(source, cls, img),
            os.path.join(val, cls, img)
        )

    for img in images[val_end:]:
        shutil.copy2(
            os.path.join(source, cls, img),
            os.path.join(test, cls, img)
        )

```

```python
for folder in ["train", "val", "test"]:
    print("\n", folder.upper())
    for cls in sorted(os.listdir(f"../dataset/{folder}")):
        print(cls, ":", len(os.listdir(f"../dataset/{folder}/{cls}")))
```
